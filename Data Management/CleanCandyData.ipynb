Task 1: Convert the GoingOut? Column into Boolean values.  If the value was left blank, assume the answer is false.


```python
candy['Q1: GOING OUT?'].fillna(0, inplace=True)
candy['Q1: GOING OUT?'].replace('No', 0, inplace=True)
candy['Q1: GOING OUT?'].replace('Yes', 1, inplace=True)
candy['Q1: GOING OUT?'] = candy['Q1: GOING OUT?'].astype(int)
```

Task 2: Convert the Gender column to conform to one of the following values: Male, Female, Other, Unspecified.  If the value was blank in the survey, then the data should load as â€œUnspecifiedâ€

```python
candy['Q2: GENDER'].fillna('Unspecified', inplace=True)
candy['Q2: GENDER'].where(candy['Q2: GENDER'].isin(['Male', 'Female', 'Unspecified']), 'Other', inplace=True)
```

Task 3: Make sure that ages are within the range of 13 to 110.  All responses from users who report to be less than 13 should be eliminated from the data set.  Responses from users who report ages above 110 should be loaded as â€œUnspecified.â€  Any non-numeric responses should be converted to â€œUnspecified.â€

```python
candy['Q3: AGE'].where([type(x) == int for x in candy['Q3: AGE']], 999, inplace=True)
candy.drop(candy[candy['Q3: AGE'] < 13].index, inplace=True)
candy['Q3: AGE'].where(candy['Q3: AGE'] <= 110, None, inplace=True)
```
