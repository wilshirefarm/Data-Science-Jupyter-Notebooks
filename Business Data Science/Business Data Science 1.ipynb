```python
#3
mu, sigma = 0, 5 # mean and standard deviation
q3_sample = np.random.normal(mu, sigma, 25000)
```

```python
#mean
q3_mean = q3_sample.sum() / len(q3_sample)
```

```python
#sd
sd_list = []
for i in q3_sample:
    a = (i - q3_mean) ** 2 
    sd_list.append(a)

q3_sd = math.sqrt(sum(sd_list) / (len(q3_sample)))

#n minus one or not?
```

```python
#4
X_means = []
Y_means = []
X_vars = []
Y_vars = []
covars = []

for i in range(100):
    x_mu = -5
    x_var = 20

    y_mu = 5
    y_var = 30

    q4_xy_sample = np.random.multivariate_normal([-5, 5], [[20, .8], [.8, 30]], 10000)

    X = q4_xy_sample[:,0]
    Y = q4_xy_sample[:,1]

    x_mean = X.sum() / len(X)
    y_mean = Y.sum() / len(Y)

    x_var_list = []
    for i in  X:
        a = (i - x_mean) ** 2 
        x_var_list.append(a)

    x_variance = sum(x_var_list) / (len(X))

    y_var_list = []
    for i in  Y:
        a = (i - y_mean) ** 2 
        y_var_list.append(a)

    y_variance = sum(y_var_list) / (len(Y))

    covar_list = []
    for i in range(len(q4_xy_sample)):
        a = (X[i] - x_mean)
        b = (Y[i] - y_mean)
        c = a * b
        covar_list.append(c)

    covar = sum(covar_list) / (len(covar_list))

    covars.append(covar)
    X_means.append(x_mean)
    Y_means.append(y_mean)
    X_vars.append(x_variance)
    Y_vars.append(y_variance)
      
print('Covariance: ' + str(np.asarray(covars).mean()))
print('Mean of X: ' + str(np.asarray(X_means).mean()))
print('Mean of Y: ' + str(np.asarray(Y_means).mean()))
print('Variance of X: ' + str(np.asarray(X_vars).mean()))
print('Variance of Y: ' + str(np.asarray(Y_vars).mean()))
print('We decided to run 100 iterations of the calculations because the covariance varied. Most of the covariances weren\'t close to .8, but if we run it mutiple times and find the mean, it is around .8')
```

```python
#3
n = 100
p = 2/3
k = 50

def binompdf(k, n, p):
    combinations = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    probability = p ** k * (1 - p) ** (n - k)
    return combinations * probability

total = 0
for i in range(k + 1):
    #print(str(i) + " " + str(binompdf(i, n, p)))
    total = total + binompdf(i, n, p)
print('The probability of getting 50 or fewer heads is: ' + str(total))
print()
print('We use the binompdf function to find the discrete probability of getting exactly k successes, and then we get \
a sum \n of all of those probabilities from 0 to 50.')



#z = st.norm.ppf(total)
#mu = n * p
#std = math.sqrt(n * p * (1-p))
#z_score = (k - mu)/std
#print(z_score)
#print(st.norm.cdf(z_score))
```
