```python
from keras.datasets import cifar10
# downloading and separating data into x_train, y_train, x_test, and y_test
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
x_train[0]
```

```python
# print out the shape our data
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
# Create data generator object
datagen = ImageDataGenerator()

# Loop through each image, horizontal flip the image, and add to new list of images
x_train_transform = []
for i in range(len(x_train)):
  x_train_image = datagen.apply_transform(x_train[i], {'flip_horizontal':True})
  x_train_transform.append(x_train_image)
```

```python
# Original first 4 images
f, axes = pyplot.subplots(2, 2)
axes[0, 0].imshow(x_train[0])
axes[0, 1].imshow(x_train[1])
axes[1, 0].imshow(x_train[2])
axes[1, 1].imshow(x_train[3])
```

```python
# Flipped first 4 images
f, axes = pyplot.subplots(2, 2)
axes[0, 0].imshow(x_train_transform[0])
axes[0, 1].imshow(x_train_transform[1])
axes[1, 0].imshow(x_train_transform[2])
axes[1, 1].imshow(x_train_transform[3])
```

```python
# Combining original images with flipped images
new_x_train = np.append(x_train, x_train_transform, axis = 0)
new_x_train.shape
```

```python
# Double the y values too
new_y_train = np.append(y_train, y_train, axis = 0)
new_y_train.shape
```

# Task 2
Try to build a Neural Network model, train on the features and report the accuracy.
Report your observations on the time taken on GPU and TPUs. 



1.   Create a CNN based model with 5 hidden layers and 100 hidden units each layer. 
2.   Create an LSTM based model with 2 hidden layers and 1024 hidden units in each layer. 



```python
# Method to create CNN with 5 hidden layers and 100 hidden units at each layer
def create_CNNmodel():
  # sequential model: allows us to create a linear stack of layers
  model = tf.keras.models.Sequential()
  
  # layer 1
  # normalize the imputs
  model.add(tf.keras.layers.BatchNormalization(input_shape=x_train.shape[1:]))
  # creating 100 channels using a 3 by 3 matrix and keeping the out matrix the same shape as the output matrix
  model.add(tf.keras.layers.Conv2D(100, (3, 3), padding='same', activation='elu'))
  # max pooling with a 2 by 2 matrix
  model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
  # drop random nodes
  model.add(tf.keras.layers.Dropout(0.25))

  # layer 2
  model.add(tf.keras.layers.BatchNormalization(input_shape=x_train.shape[1:]))
  model.add(tf.keras.layers.Conv2D(100, (3, 3), padding='same', activation='elu'))
  model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
  model.add(tf.keras.layers.Dropout(0.25))
  
  # layer 3
  model.add(tf.keras.layers.BatchNormalization(input_shape=x_train.shape[1:]))
  model.add(tf.keras.layers.Conv2D(100, (3, 3), padding='same', activation='elu'))
  model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
  model.add(tf.keras.layers.Dropout(0.25))
  
  # layer 4
  model.add(tf.keras.layers.BatchNormalization(input_shape=x_train.shape[1:]))
  model.add(tf.keras.layers.Conv2D(100, (3, 3), padding='same', activation='elu'))
  model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
  model.add(tf.keras.layers.Dropout(0.25))
  
  # layer 5
  model.add(tf.keras.layers.BatchNormalization(input_shape=x_train.shape[1:]))
  model.add(tf.keras.layers.Conv2D(100, (3, 3), padding='same', activation='elu'))
  model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
  model.add(tf.keras.layers.Dropout(0.25))
  
  # flatten the dataset
  model.add(tf.keras.layers.Flatten())
  # creating a fully connected layer with 10 outputs
  model.add(tf.keras.layers.Dense(10))
  # compare the outputs to our activation function - softmax
  model.add(tf.keras.layers.Activation('softmax'))
  return model
```

```python
# Create the model, compile it with the opitmizer and loss funtion
CNNmodel = create_CNNmodel()
CNNmodel.compile(
  optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3, ),
  loss='sparse_categorical_crossentropy',
  metrics=['sparse_categorical_accuracy'])
CNNmodel.summary()
```

```python
# Train the CNN model
%%time
CNNmodel.fit(
    new_x_train.astype(np.float32), new_y_train.astype(np.float32),
    epochs=10, batch_size=100,
    validation_data=(x_test.astype(np.float32), y_test.astype(np.float32))
)
```

```python
# Print out the accuracy of the model on the test set
scores = CNNmodel.evaluate(x_test, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```

```python
# Method to create the LSTM model
def create_LSTMmodel():
  # sequential model: allows us to create a linear stack of layers
  model = tf.keras.Sequential()
  # layer 1
  # sets return sequence to true to incorporate a second LSTM layer and an input shape of 1,3072 (our dataset that we flattened)
  model.add(layers.LSTM(1024, return_sequences=True, input_shape=(1, 3072)))
  
  # layer 2
  model.add(layers.LSTM(1024))
  # add a dense layer with an activiation function softmax
  model.add(layers.Dense(10, activation='softmax'))
  return model
```

```python
# Create the model, compile it with the opitmizer and loss funtion
LSTMmodel = create_LSTMmodel()
LSTMmodel.compile(
  optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3, ),
  loss='sparse_categorical_crossentropy',
  metrics=['sparse_categorical_accuracy'])
LSTMmodel.summary()
```

```python
# Have to flatten the input data into a shape that can be fed into the model
new_x_train_flatten = new_x_train.reshape(100000, 1, 3072)
x_test_flatten = x_test.reshape(10000, 1, 3072)
```

```python
# Train the LSTM model
%%time
LSTMmodel.fit(new_x_train_flatten.astype(np.float32), new_y_train.astype(np.float32),
          batch_size=1000,
          epochs=5,
          validation_data=(x_test_flatten.astype(np.float32), y_test.astype(np.float32)))
```

```python
# Print out the accuracy of the model on the test set
scores = LSTMmodel.evaluate(x_test_flatten, y_test, verbose=1)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```
